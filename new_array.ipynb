# 🏆 Latihan Praktikum 1 — Array Nilai Ujian


## 📌 Soal

Buat sebuah array bernama **`nilai_ujian`** yang berisi data berikut:

```python
[75, 88, 60, 92, 45, 70, 83, 55]
```

Kemudian:

1. Tampilkan seluruh nilai menggunakan **perulangan (`for`)**.
2. Tampilkan nomor urut setiap nilai.
3. Hitung jumlah seluruh nilai.
4. Hitung nilai rata-rata.
5. Tampilkan hasil total dan rata-rata.

---

## 💡 Contoh Output

```text
Nilai ke-1: 75
Nilai ke-2: 88
Nilai ke-3: 60
Nilai ke-4: 92
Nilai ke-5: 45
Nilai ke-6: 70
Nilai ke-7: 83
Nilai ke-8: 55

Total: 568
Rata-rata: 71.0
```

---

## 🔍 Petunjuk

Gunakan:

```python
sum()
len()
for
```

Rumus rata-rata:

```python
rata_rata = sum(nilai_ujian) / len(nilai_ujian)
```

---

## ✍️ Kerjakan pada cell code di bawah ini

```python
# Tulis program Anda di sini
```

````python
import array as arr

nilai_ujian = arr.array('i', [75, 88, 60, 92, 45, 70, 83, 55])

print("Daftar Nilai Ujian :")
for i in range(0, len(nilai_ujian)):
    print(f"Nilai ke-{i+1} : {nilai_ujian[i]}")

total = 0
for nilai in nilai_ujian:
    total += nilai

rata_rata = total / len(nilai_ujian)
print(f"\nTotal Nilai Ujian : {total}")
print(f"Rata-rata Nilai Ujian : {rata_rata:.2f}")
````
