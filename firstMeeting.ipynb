# Pengantar Python untuk Pemula dari Pascal

Selamat datang! Jika Anda sebelumnya belajar algoritma pemrograman dasar dengan Pascal, Python akan terasa lebih sederhana dan fleksibel. Berikut adalah beberapa perbedaan utama:

- **Tidak perlu deklarasi tipe**: Di Pascal, Anda harus mendeklarasikan `var x: integer;`, tapi di Python cukup `x = 10`.
- **Indentasi penting**: Blok kode (seperti dalam if atau loop) ditentukan oleh indentasi, bukan `begin...end`.
- **Sintaks lebih ringkas**: Tidak ada titik koma di akhir baris, dan banyak hal yang lebih sederhana.

Mari kita mulai dengan dasar-dasar!

## 1. Output Dasar (Print)

Di Pascal, Anda menggunakan `writeln('Hello World');`. Di Python, gunakan `print()`.

````python
# Contoh print
print("Hello World!")
print('Hello World!')

print("123456")
print(123456)
````

## 2. Variabel dan Tipe Data

Di Python, variabel tidak perlu dideklarasikan tipe. Python otomatis menentukan tipe berdasarkan nilai yang diberikan.

Tipe data dasar:
- `int`: bilangan bulat (seperti integer di Pascal)
- `float`: bilangan desimal
- `str`: string (teks)
- `bool`: boolean (True/False)

````python
# Contoh variabel
# myFirstName = "Ganendra"
# mySecondName = "Pradipa"

nama = "Ganendra Pratama"  # string
umur = 20  # integer
tinggi = 170.5  # float
mahasiswa = True  # boolean

print("Nama:", nama)
print("Umur:", umur)
print("Tinggi:", tinggi)
print("Mahasiswa:", mahasiswa)

# Cek tipe data
print("Tipe nama:", type(nama))
print("Tipe umur:", type(umur))
print("Tipe tinggi:", type(tinggi))
print("Tipe mahasiswa:", type(mahasiswa)) 
````

## 3. Operator

Operator di Python mirip dengan Pascal:
- Aritmatika: `+`, `-`, `*`, `/`, `//` (pembagian bulat), `%` (modulo)
- Perbandingan: `==`, `!=`, `<`, `>`, `<=`, `>=`
- Logika: `and`, `or`, `not` (bukan `and`, `or`, `not` seperti di Pascal)

````python
# Contoh operator aritmatika
a = 10
b = 3
c = 9
print("Penjumlahan:", a + b)
print("Pengurangan:", a - b)
print("Perkalian:", a * b)
print("Pembagian:", a / b)
print("Pembagian bulat:", a // b)
print("Modulo:", a % b)

# Perbandingan
print("a > b:", a > b)
print("a == b:", a == b)
print("a != c:", a != c)

# Logika
x = False
y = False
print("x and y:", x and y)
print("x or y:", x or y)
print("not x:", not x)
````

## 4. Struktur Kontrol: If-Else

Di Python, if-else menggunakan indentasi. Tidak ada `then` atau `begin...end`.

Sintaks:
```python
if kondisi:
    # kode jika benar
elif kondisi_lain:
    # kode jika kondisi_lain benar
else:
    # kode jika semua salah
```

````python
# Contoh if-else
nilai = 85

if nilai >= 90:
    print("Grade: A")
elif nilai >= 80:
    print("Grade: B")
elif nilai >= 70:
    print("Grade: C")
else:
    print("Grade: D")

# Contoh nested if
umur = 18
if umur >= 17:
    if umur >= 21:
        print("Bisa mengemudi dan minum alkohol")
    else:
        print("Bisa mengemudi tapi tidak boleh minum alkohol")
else:
    print("Belum bisa mengemudi")
````

````python
nilai = 80

if nilai >= 75:
    status = "Lulus"
else:
    status = "Tidak Lulus"

print("Nilai:", nilai)
print("Status:", status)
````

## 5. Loop: For dan While

- **For loop**: Untuk iterasi melalui sequence (list, range, dll.)
- **While loop**: Loop selama kondisi benar

Sintaks for:
```python
for item in sequence:
    # kode
```

Sintaks while:
```python
while kondisi:
    # kode
```

````python
print("\nFor loop dengan list:")
buah = ["apel", "jeruk", "mangga"]
for item in buah:
    print("Saya suka", item)
````

````python
buah.append("pinang")
for item in buah:
    print("Saya suka", item)
````

````python
# Contoh for loop
print("For loop dengan range:")
for i in range(1, 6):  # 1 sampai 5
    print("Iterasi ke-", i)

print("\nFor loop dengan list:")
buah = ["apel", "jeruk", "mangga"]
for item in buah:
    print("Saya suka", item)

# Contoh while loop
print("\nWhile loop:")
count = 1
while count <= 5:
    print("Count:", count)
    count += 1  # increment

# Loop dengan break
print("\nLoop dengan break:")
for i in range(10):
    if i == 5:
        break
    print(i)
````

## 6. Fungsi

Fungsi didefinisikan dengan `def`. Mirip dengan procedure di Pascal tapi bisa mengembalikan nilai.

Sintaks:
```python
def nama_fungsi(parameter):
    # kode
    return nilai  # opsional
```

````python
# Contoh fungsi sederhana
def sapa(nama):
    return "Halo, saya " + nama + "!. Salam kenal ya!"

hasil = sapa("Ganendra")
hasil2 = sapa("Dipa Ganendra")
print(hasil)
print(hasil2)
````

````python
panjang = 5
lebar = 3
luas = panjang * lebar
print("Luas persegi panjang:", luas)
````

````python
# Fungsi dengan multiple parameter
def hitung_luas_persegi(panjang, lebar):
    return panjang * lebar

luas = hitung_luas_persegi(5, 3)
luas1 = hitung_luas_persegi(10, 89)
print("Luas persegi panjang:", luas)
print("Luas persegi panjang:", luas1)
````

````python
# Fungsi tanpa return (seperti procedure)
def cetak_info(nama, umur):
    print(f"Nama: {nama}, \nUmur: {umur}")

cetak_info("Pratama", 20)
````

````python
# Contoh list
nilai_mahasiswa = [85, 90, 78, 92, 88]

print("List nilai:", nilai_mahasiswa)
print("Nilai pertama:", nilai_mahasiswa[0])  # indeks dimulai dari 0
print("Nilai terakhir:", nilai_mahasiswa[-1])  # indeks negatif untuk dari belakang

# Menambah item
nilai_mahasiswa.append(95)
print("Setelah append:", nilai_mahasiswa)

# Loop melalui list
print("Semua nilai:")
for nilai in nilai_mahasiswa:
    print(nilai)
````

````python
# List comprehension (cara Pythonic untuk membuat list)
kuadrat = [x**2 for x in range(1, 6)]
print("Kuadrat 1-5:", kuadrat)
````

## 8. Input dari Pengguna

Untuk membaca input, gunakan `input()`. Di Pascal, ini seperti `readln()`.

Sintaks:
```python
nama = input("Masukkan nama: ")
```

````python
# Contoh input
nama = input("Masukkan nama Anda: ")
umur = int(input("Masukkan umur Anda: "))  # konversi ke integer

print(f"Halo {nama}, umur Anda {umur} tahun.")

# Input dengan validasi sederhana
nilai = float(input("Masukkan nilai ujian (0-100): "))
if nilai >= 60:
    print(f"Lulus! dengan nilai {nilai}")
else:
    print(f"Belum lulus, coba lagi! Nilai Anda {nilai}")
````
