# Praktikum Array: Struktur Data (Python List)

Notebook ini berisi penjelasan singkat dan contoh kode untuk materi array (menggunakan list di Python) sesuai RPS minggu kedua mata kuliah Struktur Data.

## 1. Membuat & Mengakses Elemen Array (List) di Python
Array di Python diimplementasikan menggunakan tipe data `list`. <br>
Akses elemen array menggunakan indeks (dimulai dari 0).

```python
# Membuat array/list berisi nama mahasiswa
mahasiswa = ["Andi", "Budi", "Cici", "Dewi", "Eka"]
print(mahasiswa)
```

```python
# Mengakses elemen ke-2 (indeks 1)
print("Mahasiswa kedua:", mahasiswa[1])
```

## 2. Menambah Elemen ke Array
Gunakan `append()` untuk menambah elemen di akhir array.

```python
# Menambah nama ke array
mahasiswa.append("Fajar")
print(mahasiswa)
```

```python
mahasiswa.extend(["Gina", "Hadi"])
print(mahasiswa)
```

```python
mahasiswa.insert(0, "Ayu")
print(mahasiswa)
```

## 3. Menghapus Elemen dari Array
Gunakan `remove()` untuk menghapus elemen berdasarkan nilai.

```python
# Menghapus nama tertentu
mahasiswa.remove("Cici")
print("Daftar Mahasiswa Setelah Remove : ", mahasiswa)
```

```python
mahasiswa.pop(0)
print("Daftar Mahasiswa Setelah Pop : ", mahasiswa)
```

## 4. Mengupdate Nilai pada Array
Update elemen array dengan mengakses indeksnya.

```python
# Mengupdate nama pada indeks ke-0
print("Daftar Mahasiswa Sebelum Update : ", mahasiswa)

mahasiswa[0] = "Wijaya"
print("Daftar Mahasiswa Setelah Update : ", mahasiswa)
```

## 5. Traversal (Menelusuri Semua Elemen)
Gunakan loop untuk menampilkan semua elemen array.

```python
# Menampilkan semua nama mahasiswa
for nama in mahasiswa:
    print(nama)
```

```python
for i in range(len(mahasiswa)):
    print(f"Mahasiswa ke-{i+1}: {mahasiswa[i]}")
```

## 6. Mencari Elemen
Mencari posisi/keberadaan elemen dalam array.

```python
ukuranSepatu = [15, 8, 42, 23, 7, 34]
print("Banyak Data : ", len(ukuranSepatu))

# Linear Search (manual)
def linear_search(ukuranSepatu, target):
    for i in range(len(ukuranSepatu)):
        if ukuranSepatu[i] == target:
            return i  # kembalikan indeks
    return ("Kosong")  # tidak ditemukan

print(linear_search(ukuranSepatu, 15))  # Output: 3
print(linear_search(ukuranSepatu, 99))  # Output: -1
```

```python
# Menggunakan built-in Python
print(ukuranSepatu.index(23))   # 3 (exception jika tidak ada)
print(23 in ukuranSepatu)       # True
```

## 7. Praktikum: Input Nilai Mahasiswa dan Hitung Rata-rata
Latihan: Input nilai mahasiswa, simpan dalam list, lalu hitung rata-ratanya.

```python
# Input jumlah mahasiswa
n = int(input("Masukkan jumlah mahasiswa: "))
nilai = []

# Input nilai mahasiswa
for i in range(n):
    val = float(input(f"Masukkan nilai mahasiswa ke-{i+1}: "))
    nilai.append(val)

# Hitung rata-rata
rata_rata = sum(nilai) / len(nilai)
print("Nilai mahasiswa:", nilai)
print("Rata-rata nilai:", rata_rata)
```
