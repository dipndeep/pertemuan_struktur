## 🔍 Bagian 2 — Operasi Access (Mengambil Elemen)

**Kompleksitas: O(1)** — akses langsung ke alamat memori via indeks.

```
arr  = [10,  20,  30,  40,  50,  60,  70]
indeks [0]  [1]  [2]  [3]  [4]  [5]  [6]
       [-7] [-6] [-5] [-4] [-3] [-2] [-1]  ← indeks negatif
```

> 💡 **Ingat:** Indeks dimulai dari **0** (zero-based indexing)!


````python
arr = [10, 20, 30, 40, 50, 60, 70]
print(f"Array  : {arr}")
print(f"Indeks :  [0]  [1]  [2]  [3]  [4]  [5]  [6]")
print()

# Akses dengan indeks positif
print(f"arr[0]  = {arr[0]}   ← elemen pertama")
print(f"arr[3]  = {arr[3]}   ← elemen keempat")
print(f"arr[6]  = {arr[6]}   ← elemen terakhir")
print()

# Akses dengan indeks negatif (fitur Python)
print(f"arr[-1] = {arr[-1]}   ← elemen terakhir (indeks negatif)")
print(f"arr[-2] = {arr[-2]}   ← elemen kedua dari belakang")
print(f"arr[-7] = {arr[-7]}   ← sama dengan arr[0]")
````

### ✂️ Slicing — Mengambil Sebagian Array

Sintaks: `arr[start:stop:step]`
- `start` — indeks awal (inklusif), default 0
- `stop`  — indeks akhir (eksklusif)
- `step`  — langkah, default 1


````python
arr = [10, 20, 30, 40, 50, 60, 70]

print(f"arr         = {arr}")
print(f"arr[1:4]    = {arr[1:5]}       ← indeks 1, 2, 3, 4")
print(f"arr[:3]     = {arr[:3]}        ← tiga elemen pertama")
print(f"arr[4:]     = {arr[4:]}        ← dari indeks 4 sampai akhir")
print(f"arr[::2]    = {arr[::2]}       ← setiap 2 langkah")
print(f"arr[::-1]   = {arr[::-1]}      ← urutan terbalik")
print(f"arr[1:6:2]  = {arr[1:6:2]}     ← indeks 1,3,5")
````

## 🔄 Bagian 3 — Operasi Traverse

**Kompleksitas: O(n)** — setiap elemen dikunjungi tepat satu kali.

Traverse = "berjalan" melewati seluruh elemen array dari awal ke akhir.
Digunakan untuk menampilkan, menghitung, atau memfilter elemen.


````python
nilai = [85, 90, 78, 92, 88]
print(f"Array nilai: {nilai}")
print(f"Panjang array: {len(nilai)}")
print()

# Cara 1: for-each (paling Pythonic)
print("[ Cara 1: for-each ]")
for n in nilai:
    print(f"  {n}", end=",")
print()

# Cara 2: menggunakan indeks
print("\n[ Cara 2: menggunakan indeks ]")
for i in range(len(nilai)):
    print(f"  nilai[{i}] = {nilai[i]}")

# Cara 3: enumerate (indeks + nilai sekaligus) — paling informatif
print("\n[ Cara 3: enumerate — DIREKOMENDASIKAN ]")
for i, n in enumerate(nilai):
    status = "✓ LULUS" if n >= 80 else "✗ TIDAK LULUS"
    print(f"  Mahasiswa ke-{i+1}: {n} → {status}")
````

````python
# Traverse untuk menghitung statistik MANUAL (tanpa sum/min/max)
nilai = [85, 90, 78, 92, 88]

total    = 0
tertinggi = nilai[0]
terendah  = nilai[0]

for n in nilai:
    total += n
    if n > tertinggi:
        tertinggi = n
    if n < terendah:
        terendah = n

print(f"Total           : {total}")
print(f"Rata-rata       : {total / len(nilai):.2f}")
print(f"Nilai tertinggi : {tertinggi}")
print(f"Nilai terendah  : {terendah}")
````

## ➕ Bagian 4 — Operasi Insert (Menambah Elemen)

| Metode | Posisi | Kompleksitas |
|--------|--------|--------------|
| `append(val)` | Akhir | **O(1)** ✓ paling efisien |
| `insert(i, val)` | Tengah/Awal | **O(n)** — geser elemen kanan |
| `extend([...])` | Akhir (banyak) | **O(k)** |

```
Sebelum insert(2, 99):
[10, 20, 30, 40]
 [0]  [1]  [2]  [3]

Sesudah insert(2, 99):
[10, 20, 99, 30, 40]   ← 30 dan 40 bergeser kanan
 [0]  [1]  [2]  [3]  [4]
```


````python
arr = [10, 20, 30, 40, 50]
print(f"Array awal        : {arr}")
print()

# append — tambah di akhir: O(1)
arr.append(60)
print(f"Setelah append(60)     : {arr}")

# insert — tambah di posisi tertentu: O(n)
arr.insert(2, 99)
print(f"Setelah insert(2, 99)  : {arr}")

# extend — tambah beberapa elemen sekaligus
arr.extend([70, 80])
print(f"Setelah extend([70,80]): {arr}")

# + operator — buat list baru (arr asli tidak berubah)
arr2 = arr + [100, 110]
print(f"arr + [100,110]        : {arr2}")
print(f"arr asli (tidak berubah): {arr}")
````

````python
# Simulasi insert MANUAL tanpa insert()
def insert_manual(arr, indeks, nilai):
    """Sisipkan nilai di indeks tertentu menggunakan slicing."""
    return arr[:indeks] + [nilai] + arr[indeks:]

demo = [1, 2, 3, 4, 5]
print(f"Sebelum : {demo}")
demo = insert_manual(demo, 2, 99)
print(f"Sesudah insert_manual(arr, 2, 99): {demo}")
````

## ➖ Bagian 5 — Operasi Delete (Menghapus Elemen)

| Metode | Cara Hapus | Kompleksitas |
|--------|------------|--------------|
| `pop()` | Elemen terakhir | **O(1)** |
| `pop(i)` | Berdasarkan indeks | **O(n)** |
| `remove(val)` | Berdasarkan nilai (pertama) | **O(n)** |
| `del arr[i]` | Berdasarkan indeks | **O(n)** |
| `clear()` | Hapus semua | **O(1)** |


````python
arr = ['A', 'B', 'C', 'D', 'E', 'F']
print(f"Array awal: {arr}")
print()

# pop() — hapus & kembalikan elemen terakhir: O(1)
arr_test = arr.copy()
elemen = arr_test.pop()
print(f"pop()        → hapus '{elemen}', sisa: {arr_test}")

# pop(i) — hapus elemen di indeks tertentu: O(n)
arr_test = arr.copy()
elemen = arr_test.pop(1)
print(f"pop(1)       → hapus '{elemen}', sisa: {arr_test}")

# remove — hapus berdasarkan nilai (kemunculan pertama): O(n)
arr_test = arr.copy()
arr_test.remove('C')
print(f"remove('C')  → sisa: {arr_test}")

# del — hapus berdasarkan indeks
arr_test = arr.copy()
del arr_test[2]
print(f"del arr[2]   → sisa: {arr_test}")

# del slice — hapus beberapa sekaligus
arr_test = arr.copy()
del arr_test[1:3]
print(f"del arr[1:3] → sisa: {arr_test}")

# clear — hapus semua elemen
arr_test = arr.copy()
arr_test.clear()
print(f"clear()      → sisa: {arr_test}")
````

## 🔎 Bagian 6 — Operasi Search (Pencarian)

### Linear Search vs Binary Search

| | Linear Search | Binary Search |
|--|--|--|
| **Kompleksitas** | O(n) | O(log n) |
| **Syarat array** | Bebas | Harus **terurut** |
| **Cara kerja** | Cek satu per satu | Bagi dua area pencarian |
| **Cocok untuk** | Array kecil/tidak terurut | Array besar & terurut |


````python
# ── LINEAR SEARCH ──
def metode_linear(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

data = [15, 8, 42, 23, 7, 34, 19, 56]
print(f"Data: {data}")
print()

for cari in [23, 42, 99, 15]:
    idx = metode_linear(data, cari)
    if idx != -1:
        print(f"  Mencari {cari} → ✓ ditemukan di indeks [{idx}]")
    else:
        print(f"  Mencari {cari} → ✗ tidak ditemukan")

print()
# Built-in Python
print(f"data.index(23)  = {data.index(23)}")
print(f"23 in data      = {23 in data}")
print(f"99 in data      = {99 in data}")
````

````python
ukuranSepatu = [15, 8, 42, 23, 7, 34]
print("Banyak Data : ", len(ukuranSepatu))

# Linear Search (manual)
def linear_search(ukuranSepatu, target):
    for i in range(len(ukuranSepatu)):
        if ukuranSepatu[i] == target:
            return i  # kembalikan indeks
    return ("Kosong")  # tidak ditemukan

print(linear_search(ukuranSepatu, 15))  # Output: 3
print(linear_search(ukuranSepatu, 99))  # Output: -1
````

````python
# ── BINARY SEARCH — hanya untuk array TERURUT ──
def binary_search(arr, target):
    kiri  = 0
    kanan = len(arr) - 1
    while kiri <= kanan:
        tengah = (kiri + kanan) // 2
        print(f"    Periksa indeks [{tengah}] = {arr[tengah]}", end="")
        if arr[tengah] == target:
            print(f" ← DITEMUKAN! ✓")
            return tengah
        elif arr[tengah] < target:
            print(f" → target {target} lebih besar, geser kiri ke [{tengah+1}]")
            kiri = tengah + 1
        else:
            print(f" → target {target} lebih kecil, geser kanan ke [{tengah-1}]")
            kanan = tengah - 1
    return ("Ga ada kocak")

data_ini = [15, 8, 42, 23, 7, 34, 19, 56]
terurut = sorted(data_ini)
print(f"Indeks Awal: 0")
print(f"Indeks Akhir: {len(data_ini) - 1}")
print(f"Data terurut: {terurut}")
print(f"\nMencari 23:")
idx = binary_search(terurut, 23)
print(f"Hasil: indeks [{idx}]")
````

## 🏁 Bagian 8 — Array Multidimensi (2D)

Array 2D = array yang elemennya juga array → seperti **tabel/matriks**.

```
         kolom [0]  [1]  [2]
baris [0]  →  [ 1,   2,   3 ]
baris [1]  →  [ 4,   5,   6 ]
baris [2]  →  [ 7,   8,   9 ]

Akses: matrix[baris][kolom]
Contoh: matrix[1][2] = 6
```


````python
# Deklarasi matrix 3x3
matrix = [
    [1, 2, 3],   # baris 0
    [4, 5, 6],   # baris 1
    [7, 8, 9]    # baris 2
]

print("Matrix 3x3:")
for baris in matrix:
    print(f"  {baris}")
print()

# Akses elemen
print(f"matrix[0][0] = {matrix[0][0]}  ← baris 0, kolom 0")
print(f"matrix[1][1] = {matrix[1][1]}  ← baris 1, kolom 1 (tengah)")
print(f"matrix[2][2] = {matrix[2][2]}  ← baris 2, kolom 2")
print(f"matrix[1][2] = {matrix[1][2]}  ← baris 1, kolom 2")
````

````python
# Traverse array 2D dengan nested loop
matrix = [[1,2,3],[4,5,6],[7,8,9]]

print("Traverse matrix 2D:")
for i in range(len(matrix)):
    for j in range(len(matrix[i])):
        print(f"  matrix[{i}][{j}] = {matrix[i][j]}")
````

````python
# Studi Kasus: Tabel Nilai Mahasiswa
# Baris = mahasiswa, Kolom = mata kuliah

nama  = ["Alice", "Bob", "Charlie"]
mapel = ["Matematika", "Struktur Data", "Basis Data", "Jaringan"]

nilai_mhs = [
    [85, 90, 78, 92],  # Alice
    [75, 88, 95, 80],  # Bob
    [92, 70, 85, 88],  # Charlie
]

# Tampilkan tabel
print(f"{'Nama':<12}", end="")
for mk in mapel:
    print(f"{mk:<16}", end="")
print("Rata-rata")
print("-" * 90)

for i, nama_s in enumerate(nama):
    print(f"{nama_s:<12}", end="")
    total = 0
    for j in range(len(mapel)):
        print(f"{nilai_mhs[i][j]:<16}", end="")
        total += nilai_mhs[i][j]
    print(f"{total/len(mapel):.1f}")
````

## 🎓 Bagian 10 — Studi Kasus: Sistem Nilai Sederhana

Kita akan menggabungkan semua operasi array dalam satu studi kasus nyata.


````python
def sistem_nilai(daftar_nilai, daftar_nama=None):
    if not daftar_nilai:
        return {"error": "Array kosong"}

    n     = len(daftar_nilai)
    total = sum(daftar_nilai)
    lulus = [v for v in daftar_nilai if v >= 60]

    return {
        "jumlah"      : n,
        "total"       : total,
        "rata_rata"   : round(total / n, 2),
        "nilai_maks"  : max(daftar_nilai),
        "nilai_min"   : min(daftar_nilai),
        "lulus"       : len(lulus),
        "tidak_lulus" : n - len(lulus),
        "pct_lulus"   : round(len(lulus) / n * 100, 1),
    }

# ── Contoh Penggunaan ──
nilai_kelas = [85, 72, 90, 45, 88, 63, 77, 55, 92, 68]
nama_kelas  = ["A","B","C","D","E","F","G","H","I","J"]

print(f"Data nilai : {nilai_kelas}")
print(f"Nama       : {nama_kelas}")
print()

stat = sistem_nilai(nilai_kelas)
print("[ STATISTIK KELAS ]")
for k, v in stat.items():
    print(f"  {k:<15}: {v}")

# Peringkat
sorted_idx = sorted(range(len(nilai_kelas)), key=lambda i: nilai_kelas[i],  reverse=True)
print("\n[ PERINGKAT ]")
for rank, idx in enumerate(sorted_idx, 1):
    nama_s = nama_kelas[idx]
    nilai  = nilai_kelas[idx]
    status = "LULUS" if nilai >= 60 else "TIDAK LULUS"
    print(f"  #{rank:<3} {nama_s}: {nilai:>3}  [{status}]")
````
