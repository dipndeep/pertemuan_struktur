## 🧩 BAGIAN 1 — Konsep Dasar: Node & Pointer

### Apa itu Node?
**Node** adalah unit terkecil dari linked list. Setiap node menyimpan dua hal:
- **Data** → nilai yang disimpan
- **Pointer (next)** → referensi ke node berikutnya

```
┌──────────┬──────────┐      ┌──────────┬──────────┐      ┌──────────┬──────────┐
│  Data: 1 │  next ───┼─────►│  Data: 2 │  next ───┼─────►│  Data: 3 │  next:   │
│          │          │      │          │          │      │          │  None    │
└──────────┴──────────┘      └──────────┴──────────┘      └──────────┴──────────┘
     Node 1                       Node 2                       Node 3 (tail)
  (head)
```

### Apa itu Pointer?
Dalam Python, **pointer** direpresentasikan sebagai referensi objek. Variabel `next` pada setiap node menyimpan *alamat memori* (referensi) node berikutnya, bukan nilainya langsung.

````python
# ============================================
# KONSEP NODE
# Membuat class Node sederhana
# ============================================

class Node:
    def __init__(self, data):
        self.data = data   # menyimpan nilai/data
        self.next = None   # pointer ke node berikutnya (awalnya None)

# --- Membuat node secara manual ---
node1 = Node(10)
node2 = Node(20)
node3 = Node(30)

# Menghubungkan node menggunakan pointer
node1.next = node2   # node1 menunjuk ke node2
node2.next = node3   # node2 menunjuk ke node3
# node3.next tetap None → ini adalah node terakhir (tail)

print("=== VISUALISASI NODE ===")
print(f"node1.data = {node1.data} | node1.next → {node1.next.data}")
print(f"node2.data = {node2.data} | node2.next → {node2.next.data}")
print(f"node3.data = {node3.data} | node3.next → {node3.next}  (None = akhir list)")

# print("\n=== CEK POINTER/REFERENSI ===")
# print(f"Alamat memori node2     : {id(node2)}")
# print(f"Alamat yang ditunjuk node1.next: {id(node1.next)}")
# print(f"Apakah node1.next adalah node2? → {node1.next is node2}")
````

## 📚 BAGIAN 2 — Jenis-Jenis Linked List

| Jenis | Deskripsi | Pointer |
|---|---|---|
| **Singly Linked List** | Setiap node hanya punya 1 pointer (ke depan) | `next` |
| **Doubly Linked List** | Setiap node punya 2 pointer (depan & belakang) | `next`, `prev` |
| **Circular Linked List** | Node terakhir menunjuk kembali ke node pertama | `next` (memutar) |

Pada pertemuan ini kita fokus pada **Singly Linked List**.

---
## 🛠️ BAGIAN 3 — Implementasi Singly Linked List

### Operasi yang akan diimplementasikan:
1. `append(data)` — tambah node di akhir
2. `prepend(data)` — tambah node di awal
3. `insert_after(target, data)` — sisipkan node setelah node tertentu
4. `delete(data)` — hapus node berdasarkan nilai
5. `search(data)` — cari node berdasarkan nilai
6. `display()` — tampilkan semua isi linked list
7. `length()` — hitung jumlah node

````python
# ============================================
# IMPLEMENTASI SINGLY LINKED LIST LENGKAP
# ============================================

class Node:
    """Representasi satu node dalam linked list."""
    def __init__(self, data):
        self.data = data
        self.next = None


class SinglyLinkedList:
    """Implementasi Singly Linked List."""

    def __init__(self):
        self.head = None  # awal list, awalnya kosong

    # ── 1. DISPLAY ────────────────────────────────────
    def display(self):
        """Tampilkan semua elemen dalam linked list."""
        if self.head is None:
            print("List kosong!")
            return

        current = self.head
        result = []
        while current is not None:
            result.append(str(current.data))
            current = current.next  # pindah ke node berikutnya

        print(" → ".join(result) + " → None")

    # ── 2. APPEND (tambah di akhir) ────────────────────
    def append(self, data):
        """Tambah node baru di akhir list."""
        new_node = Node(data)

        if self.head is None:       # jika list kosong
            self.head = new_node
            return

        # Traversal: cari node terakhir
        current = self.head
        while current.next is not None:
            current = current.next

        current.next = new_node     # hubungkan ke node baru

    # ── 3. PREPEND (tambah di awal) ────────────────────
    def prepend(self, data):
        """Tambah node baru di awal list."""
        new_node = Node(data)
        new_node.next = self.head   # node baru menunjuk ke head lama
        self.head = new_node        # head sekarang adalah node baru

    # ── 4. INSERT AFTER ────────────────────────────────
    def insert_after(self, target_data, new_data):
        """Sisipkan node baru setelah node dengan nilai target_data."""
        current = self.head
        while current is not None:
            if current.data == target_data:
                new_node = Node(new_data)
                new_node.next = current.next  # sambung ke node setelah target
                current.next = new_node       # target menunjuk ke node baru
                return
            current = current.next

        print(f"Data '{target_data}' tidak ditemukan dalam list.")

    # ── 5. DELETE ──────────────────────────────────────
    def delete(self, data):
        """Hapus node pertama yang memiliki nilai data."""
        if self.head is None:
            print("List kosong, tidak ada yang dihapus.")
            return

        # Kasus: node yang dihapus adalah head
        if self.head.data == data:
            self.head = self.head.next
            print(f"Node dengan data '{data}' berhasil dihapus.")
            return

        # Cari node sebelum node yang akan dihapus
        current = self.head
        while current.next is not None:
            if current.next.data == data:
                current.next = current.next.next  # lewati node yang dihapus
                print(f"Node dengan data '{data}' berhasil dihapus.")
                return
            current = current.next

        print(f"Data '{data}' tidak ditemukan.")

    # ── 6. SEARCH ──────────────────────────────────────
    def search(self, data):
        """Cari node berdasarkan nilai, kembalikan posisi (0-indexed)."""
        current = self.head
        position = 0

        while current is not None:
            if current.data == data:
                print(f"Data '{data}' ditemukan di posisi ke-{position}")
                return position
            current = current.next
            position += 1

        print(f"Data '{data}' tidak ditemukan dalam list.")
        return -1

    # ── 7. LENGTH ──────────────────────────────────────
    def length(self):
        """Hitung jumlah node dalam list."""
        count = 0
        current = self.head
        while current is not None:
            count += 1
            current = current.next
        return count


print("✅ Class SinglyLinkedList berhasil didefinisikan!")
````

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    # ── 1. APPEND ────────────────────────────────────
    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        current = self.head
        while current.next is not None:
            current = current.next
        current.next = new_node

    # ── 2. DISPLAY ────────────────────────────────────
    def display(self):
        if self.head is None:
            print("List kosong!")
            return
        current = self.head
        result = []
        while current is not None:
            result.append(str(current.data))
            current = current.next
        print(" → ".join(result) + " → None")

    # ── 3. PREPEND ────────────────────────────────────
    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # ── 4. LENGTH ────────────────────────────────────
    def length(self):
        count = 0
        current = self.head
        while current is not None:
            count += 1
            current = current.next
        return count

print("=" * 45)
print("📌 DEMO 2 — PREPEND (Tambah di Awal)")
print("=" * 45)

# DEMO
ll = SinglyLinkedList()
ll.append(10)
ll.append(20)
ll.append(30)
ll.append(40)

print("\nSebelum prepend:")
ll.display()
print(f"Panjang list: {ll.length()} node")

ll.prepend(5)

print("\nSetelah prepend(5):")
ll.display()
print(f"Panjang list: {ll.length()} node")
````

````python
# ============================================
# DEMO: INSERT AFTER
# ============================================

print("=" * 45)
print("📌 DEMO 3 — INSERT AFTER (Sisip di Tengah)")
print("=" * 45)

print("\nSebelum insert:")
ll.display()

ll.insert_after(20, 25)  # sisipkan 25 setelah node yang berisi 20

print("\nSetelah insert_after(20, 25):")
ll.display()
print(f"Panjang list: {ll.length()} node")
````

````python
# ============================================
# DEMO: SEARCH
# ============================================

print("=" * 45)
print("📌 DEMO 4 — SEARCH (Pencarian)")
print("=" * 45)

print("\nIsi list saat ini:")
ll.display()

print()
ll.search(25)   # ada di tengah
ll.search(5)    # ada di awal
ll.search(40)   # ada di akhir
ll.search(99)   # tidak ada
````

````python
# ============================================
# DEMO: DELETE
# ============================================

print("=" * 45)
print("📌 DEMO 5 — DELETE (Hapus Node)")
print("=" * 45)

print("\nIsi list sebelum delete:")
ll.display()

print()
ll.delete(5)    # hapus node pertama (head)
ll.display()

print()
ll.delete(25)   # hapus node di tengah
ll.display()

print()
ll.delete(40)   # hapus node terakhir (tail)
ll.display()

print()
ll.delete(99)   # data tidak ada

print(f"\nPanjang akhir list: {ll.length()} node")
````

## 🔄 BAGIAN 5 — Analisis Kompleksitas Waktu

| Operasi | Kompleksitas | Keterangan |
|---|---|---|
| `append` | O(n) | Harus traversal ke akhir |
| `prepend` | **O(1)** | Langsung ganti head |
| `insert_after` | O(n) | Harus cari node target |
| `delete` | O(n) | Harus cari node target |
| `search` | O(n) | Traversal dari head |
| `length` | O(n) | Traversal seluruh list |

> 💡 **Catatan:** Operasi `prepend` lebih efisien daripada `append` karena tidak perlu traversal.

````python
# ============================================
# VISUALISASI TRAVERSAL (Step by Step)
# ============================================

print("=" * 50)
print("🔍 VISUALISASI TRAVERSAL STEP BY STEP")
print("=" * 50)

# Buat list baru untuk demo
demo = SinglyLinkedList()
for val in ["Alice", "Bob", "Charlie", "Diana"]:
    demo.append(val)

print("\nIsi list:")
demo.display()

print("\nProses traversal mencari 'Charlie':")
current = demo.head
step = 0
while current is not None:
    status = "✅ KETEMU!" if current.data == "Charlie" else "⏭️  lewati"
    print(f"  Langkah {step}: node.data = '{current.data}' → {status}")
    if current.data == "Charlie":
        break
    current = current.next
    step += 1
````

## 💼 BAGIAN 6 — Studi Kasus: Antrian Mahasiswa

Gunakan Linked List untuk mensimulasikan **antrian bimbingan dosen** di Jurusan Sistem Informasi.

````python
# ============================================
# STUDI KASUS: ANTRIAN BIMBINGAN DOSEN
# ============================================

class AntriMahasiswa:
    """Sistem antrian bimbingan dosen menggunakan Singly Linked List."""

    def __init__(self):
        self.antrian = SinglyLinkedList()

    def daftar(self, nama_mahasiswa):
        """Mahasiswa mendaftar antrian (masuk di akhir)."""
        self.antrian.append(nama_mahasiswa)
        print(f"✅ {nama_mahasiswa} berhasil mendaftar antrian.")

    def panggil(self):
        """Panggil mahasiswa pertama dalam antrian."""
        if self.antrian.head is None:
            print("❌ Antrian kosong, tidak ada mahasiswa yang menunggu.")
            return

        nama = self.antrian.head.data
        self.antrian.delete(nama)
        print(f"🔔 Memanggil: {nama} — silakan masuk ruangan!")

    def lihat_antrian(self):
        """Tampilkan seluruh antrian."""
        print("\n📋 Antrian saat ini:")
        self.antrian.display()
        print(f"   Total: {self.antrian.length()} mahasiswa")


# --- Simulasi ---
print("=" * 50)
print("  SISTEM ANTRIAN BIMBINGAN DOSEN")
print("  Jurusan Sistem Informasi")
print("=" * 50)

sistem = AntriMahasiswa()

# Mahasiswa mendaftar
print("\n--- Mahasiswa mendaftar ---")
sistem.daftar("Budi Santoso")
sistem.daftar("Siti Rahayu")
sistem.daftar("Andi Wijaya")
sistem.daftar("Rina Kusuma")

# Lihat antrian
sistem.lihat_antrian()

# Dosen memanggil satu per satu
print("\n--- Dosen memanggil mahasiswa ---")
sistem.panggil()
sistem.panggil()

# Lihat antrian setelah 2 dipanggil
sistem.lihat_antrian()

# Mahasiswa baru masuk antrian
print("\n--- Mahasiswa baru mendaftar ---")
sistem.daftar("Doni Firmansyah")
sistem.lihat_antrian()
````
