# 1. LIST

```python
nilaiMahasiswa = [85, 90, 78, 92, 88, 95, 92]

# --- READ DATA ---
print("Nilai mahasiswa:", nilaiMahasiswa)
print(f"Nilai mahasiswa: {nilaiMahasiswa}")

# --- ADD DATA ---
print(f"\nNilai mahasiswa sebelum ditambahkan: {nilaiMahasiswa}")
nilaiMahasiswa.append(91)
print(f"Nilai mahasiswa setelah append: {nilaiMahasiswa}")
nilaiMahasiswa.extend([89, 87])
print(f"Nilai mahasiswa setelah extend: {nilaiMahasiswa}")
nilaiMahasiswa.insert(2, 80)
print(f"Nilai mahasiswa setelah insert: {nilaiMahasiswa}")

# --- UPDATE DATA ---
print(f"\nNilai mahasiswa sebelum update: {nilaiMahasiswa}")
nilaiMahasiswa[3] = 0
print(f"Nilai mahasiswa setelah update: {nilaiMahasiswa}")

# --- DELETE DATA ---
print(f"\nNilai mahasiswa sebelum delete: {nilaiMahasiswa}")
nilaiMahasiswa.remove(0)
print(f"Nilai mahasiswa setelah remove: {nilaiMahasiswa}")
nilaiMahasiswa.pop(2)
print(f"Nilai mahasiswa setelah pop: {nilaiMahasiswa}")
nilaiMahasiswa.pop()
print(f"Nilai mahasiswa setelah pop terakhir: {nilaiMahasiswa}")
del nilaiMahasiswa[7]
print(f"Nilai mahasiswa setelah del: {nilaiMahasiswa}")
```

# 2. TUPLE

```python
ukuranSepatu = (40, 41, 42, 43, 44, 42, 50, 41)
print(f"Ukuran sepatu: {ukuranSepatu}")

# --- READ DATA ---
print(f"\nUkuran sepatu: {ukuranSepatu}")

# --- COUNT DATA ---
print(f"\nJumlah ukuran sepatu 42         : {ukuranSepatu.count(42)}")

# --- INDEX DATA ---
print(f"\nUkuran sepatu pada indeks 1     : {ukuranSepatu[1]}")
print(f"Indeks pertama ukuran sepatu 41 : {ukuranSepatu.index(41)}")
```

# 3. SET

```python
umur = {20, 25, 30, 35, 40, 25, 30}

# --- READ DATA ---
print(f"Umur mahasiswa: {umur}")

# --- ADD DATA ---
print(f"\nUmur mahasiswa sebelum ditambahkan: {umur}")
umur.add(45)
print(f"Umur mahasiswa setelah add: {umur}")
umur.update([50, 55, 60])
print(f"Umur mahasiswa setelah update: {umur}")

# --- DELETE DATA ---
print(f"\nUmur mahasiswa sebelum delete: {umur}")
umur.remove(25)
print(f"Umur mahasiswa setelah remove: {umur}")
umur.discard(30)
print(f"Umur mahasiswa setelah discard: {umur}")
umur.pop()
print(f"Umur mahasiswa setelah pop: {umur}")
```

# 4. DICTIONARY

```python
dataMahasiswa = {
    "nama": "Feri Stefanus",
    "jurusan": "Sistem Informasi",
}

# --- READ DATA ---
print(f"Data mahasiswa    : {dataMahasiswa}")
print(f"Nama mahasiswa    : {dataMahasiswa['nama']}")
print(f"Jurusan mahasiswa : {dataMahasiswa['jurusan']}")

# --- ADD DATA ---
dataMahasiswa["semester"] = 4
print(f"\nData mahasiswa setelah ditambahkan semester: {dataMahasiswa}")
print(f"Nama mahasiswa    : {dataMahasiswa['nama']}")
print(f"Jurusan mahasiswa : {dataMahasiswa['jurusan']}")
print(f"Semester          : {dataMahasiswa['semester']}")

# --- UPDATE DATA 1 KEY---
dataMahasiswa["semester"] = 10
print(f"\nData mahasiswa setelah update semester: {dataMahasiswa}")
print(f"Nama mahasiswa    : {dataMahasiswa['nama']}")
print(f"Jurusan mahasiswa : {dataMahasiswa['jurusan']}")
print(f"Semester          : {dataMahasiswa['semester']}")

# --- UPDATE DATA > 1 KEY ---
dataMahasiswa.update({"nama": "Stefanus Cahya", "jurusan": "Teknik Informatika"})
print(f"\nData mahasiswa setelah update nama dan jurusan: {dataMahasiswa}")
print(f"Nama mahasiswa    : {dataMahasiswa['nama']}")
print(f"Jurusan mahasiswa : {dataMahasiswa['jurusan']}")
print(f"Semester          : {dataMahasiswa['semester']}")
```
