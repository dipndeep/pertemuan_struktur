## **1. Input dan Operasi Data**

```python
nama = input("Masukkan nama: ")
umur = int(input("Masukkan umur (tahun): "))
tinggi = float(input("Masukkan tinggi (centimeter): "))

umur_bulan = umur * 12
tinggi_cm = tinggi * 100

print("\n=== Data Diri ===")
print("Nama:", nama)
print("Umur:", umur, "tahun")
print("Umur dalam bulan:", umur_bulan, "bulan")
print("Tinggi:", tinggi_cm, "cm")
```

## **2. Percabangan (If-Else)**

```python
nilai = int(input("Masukkan nilai: "))

if nilai >= 85:
    grade = "A"
elif nilai >= 70:
    grade = "B"
elif nilai >= 60:
    grade = "C"
elif nilai >= 50:
    grade = "D"
else:
    grade = "E"

print("Nilai huruf:", grade)
```

## **3. Perulangan (SistemInformasi)**

```python
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("SistemInformasi")
    elif i % 3 == 0:
        print("Sistem")
    elif i % 5 == 0:
        print("Informasi")
    else:
        print(i)
```

## **4. List (Array Sederhana)**

```python
data = []

for i in range(5):
    angka = int(input(f"Masukkan angka ke-{i+1}: "))
    data.append(angka)

print("Data:", data)
print("Nilai terbesar:", max(data))
print("Nilai terkecil:", min(data))
print("Rata-rata:", sum(data)/len(data))
```

## **5. Manipulasi List (Genap & Ganjil)**

```python
data = list(map(int, input("Masukkan angka (pisahkan dengan spasi): ").split()))

genap = []
ganjil = []

for angka in data:
    if angka % 2 == 0:
        genap.append(angka)
    else:
        ganjil.append(angka)

print("Angka genap:", genap)
print("Angka ganjil:", ganjil)
```
